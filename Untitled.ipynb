```python
import os
import math
import random
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
from numpy.linalg import inv
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.datasets import make_regression
from scipy import signal



#Creating the linear signal with noise


X, y = make_regression(n_samples=100, n_features=1, noise=10)

# OLS 

b = inv(X.T.dot(X)).dot(X.T).dot(y)
print(b)
# predict using coefficients
yhat = X.dot(b)
# plot data and predictions
plt.scatter(X, y)
plt.plot(X, yhat, color='red')
plt.show()

#Create a square plot with noise to filter using non linear regression/ LMS
sample = 200
t = np.linspace(0, 1, 200, endpoint=True)
pureSquareSignal = signal.square(30*t)
noise = 0.0008*np.asarray(random.sample(range(0,200),sample))
noisySignal = pureSquareSignal + noise
plt.plot(noisySignal)




```
